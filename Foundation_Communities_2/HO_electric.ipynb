```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("bills_Homestead_Oaks_HO_Electric.xls")
```

```python
df.head(10)
```

```python
df.drop('Fuel',axis=1,inplace=True)
```

```python
df['Total_Bill']=df['Bill Amount']+df['Third Party Bill']
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['StartYear'], df['StartMonth'] = df['Start Date'].dt.year, df['Start Date'].dt.month
```

```python
df['End Date'] = pd.to_datetime(df['End Date'])
df['EndYear'], df['EndMonth'] = df['End Date'].dt.year, df['End Date'].dt.month
```

```python
Groups=df.groupby('StartYear')['Usage'].sum()
```

```python
df.shape
```

```python
df.head(10)
```

```python
x=df['StartYear']
y=df['Usage']
plt.bar(x,y,color='Green',width=0.5)
plt.title('Graph plotted between the usage and months')
plt.xlabel('Years')
plt.ylabel('Usage')
plt.show()
```

```python
x=df['StartYear']
y=df['Total_Bill']
plt.bar(x,y,color='Green',width=0.5)
plt.title('Graph plotted between the years and the Total_Bill')
plt.xlabel('Years')
plt.ylabel('Total Bill')
plt.show()
```

```python
x=df.groupby('StartYear')['Usage'].sum()
labels = '2015', '2016', '2017', '2018','2019','2020','2021','2022'
plt.pie(x,labels=labels)
plt.title("Graph Plotted between Usage and Start Year")
plt.show()
```

```python
x=df.groupby('StartYear')['Total_Bill'].sum()
labels = '2015', '2016', '2017', '2018','2019','2020','2021','2022'
plt.pie(x,labels=labels)
plt.title("Graph Plotted between Total_Bill and Start Year")
plt.show()
```

```python
x=df['StartYear']
y=df['Usage']
plt.title("Plot between Start Year and Usage")
plt.scatter(x,y)
plt.show()
```
