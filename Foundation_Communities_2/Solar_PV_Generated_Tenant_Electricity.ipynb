```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df=pd.read_excel("bills_Homestead_Oaks_Aggregate_Solar_PV_Generated_Tenant_Elec.xls")
```

```python
import matplotlib.pyplot as plt
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['StartYear'], df['StartMonth'] = df['Start Date'].dt.year, df['Start Date'].dt.month
```

```python
df['End Date'] = pd.to_datetime(df['End Date'])
df['EndYear'], df['EndMonth'] = df['End Date'].dt.year, df['End Date'].dt.month
```

```python
Groups=df.groupby('StartYear')['Usage'].sum()
```

```python
df.head
```

```python
x=df['StartYear']
y=df['Usage']
plt.bar(x,y,color='Green',width=0.5)
plt.title('Graph plotted between the usage and months')
plt.xlabel('Years')
plt.ylabel('Usage')
plt.show()
```

```python
x=df['StartMonth']
y=df['Usage']
plt.bar(x,y,color='Green',width=0.5)
plt.xlabel('Months')
plt.ylabel("Usage")
plt.title('Graph plotted between the usage and months')
plt.show()
```
